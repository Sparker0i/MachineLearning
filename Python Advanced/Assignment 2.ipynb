```python
def flatten(S):
    if S == []:
        return S
    if isinstance(S[0] , list):
        return flatten(S[0]) + flatten(S[1:])
    return S[:1] + flatten(S[1:])
            
def nested_sum(nestedlist):
    newList = flatten(nestedlist)
    print sum(newList)
    
nested_sum([1 , [2] , [3 , 5 , 6 , 7 , 9]])
```

```python
def isSorted(S):
    for i in range(0,len(S) - 1):
        if (S[i] > S[i + 1]):
            return False
    return True

isSorted(['a' , 'b' , 'c' , 'a'])
```

```python
def paradox(S):
    if len(S) != len(set(S)):
        return False
    else:
        return True
    
paradox([1 , 2 , 1])
```

```python
import random

def hasDuplicates(S):
    return len(S) != len(set(S))

def randomBirthdays(n):
    t = []
    for i in range(0,n):
        t.append(random.randint(1 , 365))
    return t

students = 23
if hasDuplicates(randomBirthdays(students)):
    print True
else:
    print False
```

```python
import string

def rotateLetter(letter , n):
    if letter.isupper():
        start = ord('A')
    elif letter.islower():
        start = ord('a')
    else:
        return letter
    
    c = ord(letter) - start
    i = (c + n) % 26 + start
    return chr(i)
    
def rotateWord(word , n):
    res = '';
    for letter in word:
        res += rotateLetter(letter , n)
    return res
    
def makeWordDict():
    d = dict()
    fin = open('words.txt')
    for line in fin:
        word = line.strip().lower()
        d[word] = word
    return d
    
def rotatePairs(word , wordDict):
    for i in range(1,14):
        rotated = rotateWord(word , i)
        if rotated in wordDict:
            print word , i , rotated
            
wordDict = makeWordDict()
for word in wordDict:
    rotatePairs(word , wordDict)
```

```python
def sumAll(*args):
    sum = 0
    for num in args:
        sum += num
    return sum

t = (7 , 4 , 1 , 5 , 6)
sumAll(*t)
```
